```python

def gettingstartpoints(numberofcells = 50,weightscellcycledistributions=[100,0,0],
                       lambdaG1 = 1.15,kappaG1 = 9,lambdaS = 16.9,kappaS = 128,lambdaG2 = 5.59,kappaG2 = 19,lambdaM=0,kappaM=0,
                      outputpath='',treatment='DMSO'):
    import random
    import matplotlib.pyplot as plt
    import scipy.special as sps  
    import numpy as np

    data=[]
    phases=[]
    alllengths=[]
    lengthallG1=[]
    lengthallS=[]
    lengthallG2=[]
    lengthallM=[]

    for cell in list(range(1,int(numberofcells)+1)):
        starttime=0
        startphase = random.choices(["G1",'S','G2M'],weights=weightscellcycledistributions)
        #print(startphase[0])
        lengthG1=np.random.gamma(shape=kappaG1,scale=1/lambdaG1)
        lengthS=np.random.gamma(shape=kappaS,scale=1/lambdaS)
        lengthG2=np.random.gamma(shape=kappaG2,scale=1/lambdaG2)
        lengthM=np.random.gamma(shape=kappaM,scale=1/lambdaM)
        maximallength=lengthG1+lengthS+lengthG2+lengthM
        #print("lengthG1",lengthG1)
        #print("lengthS",lengthS)
        #print("lengthG2",lengthG2)

        if startphase[0] == 'G1':
            starttime=np.random.uniform(0,lengthG1,1)
            #starttime=0
            #print(starttime[0])
        if startphase[0] == 'S':
            starttime=np.random.uniform(lengthG1,lengthG1+lengthS,1)
            #print(starttime[0])

        if startphase[0] == 'G2M':
            starttime=np.random.uniform(lengthG1+lengthS,lengthG1+lengthS+lengthG2+lengthM,1)
            #print(starttime[0])

        data +=[starttime]
        phases +=[startphase[0]]
        lengthallG1+=[lengthG1]
        lengthallS+=[lengthS]
        lengthallG2+=[lengthG2]
        lengthallM+=[lengthM]
        alllengths +=[maximallength]
        
    if treatment == 'DMSO' or treatment == 'control' or treatment == 'MQ':
        colouring = '#5B9BD5' 
    elif treatment == 'V1' or treatment == 'dTAG-V1':
        colouring = '#FF5050'
    
    fig, (ax1, ax2,ax3,ax4) = plt.subplots(1, 4,sharey=True)
    fig.suptitle(str('Modelled gamma first cycle '+treatment))
    fig.set_figwidth(10)
    ax1.hist(lengthallG1,bins=50,color=[colouring])
    ax1.set_title('G1 phase')
    ax1.text(0.45,0.98, s=str('lambda: '+str(lambdaG1)+'\nkappa: '+str(kappaG1)),
             horizontalalignment='left',verticalalignment='top',fontsize=10,transform=ax1.transAxes)
    ax1.set_xlim(0,24)
    ax2.hist(lengthallS,bins=50,color=[colouring])
    ax2.set_title('S phase')
    ax2.set_xlim(0,24)
    ax2.text(0.45,0.98, s=str('lambda: '+str(lambdaS)+'\nkappa: '+str(kappaS)),
             horizontalalignment='left',verticalalignment='top',fontsize=10,transform=ax2.transAxes)
    ax3.hist(lengthallG2,bins=50,color=[colouring])
    ax3.set_title('G2 phase')
    ax3.text(0.45,0.98, s=str('lambda: '+str(lambdaG2)+'\nkappa: '+str(kappaG2)),
             horizontalalignment='left',verticalalignment='top',fontsize=10,transform=ax3.transAxes)
    ax3.set_xlim(0,24)
    ax4.hist(lengthallM,bins=50,color=[colouring])
    ax4.set_title('M phase')
    ax4.text(0.45,0.98, s=str('lambda: '+str(lambdaM)+'\nkappa: '+str(kappaM)),
             horizontalalignment='left',verticalalignment='top',fontsize=10,transform=ax4.transAxes)
    ax4.set_xlim(0,1.5)
    fig.savefig(str(outputpath+'modelledgammadistributions_1'+treatment+'_lastmodel.png'))
    plt.close()
    
    #for i in range(0,len(alllengths)):
    #    summation=lengthallG1[i]+lengthallS[i]+lengthallG2[i]+lengthallM[i]
    #    print(alllengths[i],summation)
    return [data,phases,alllengths,lengthallG1,lengthallS,lengthallG2,lengthallM]


```

```python
modelnumber=558


numberofcells=10000
takingtimepoints=49
deltatimeinterval=1
weightscellcycledistributions=[100,0,0]
outputpath='C:/Users/Gebruiker/OneDrive - Universiteit Utrecht/CSDB/Major Internship/Modelling cell cycle/TempFigures/'
inputpath='C:/Users/Gebruiker/OneDrive - Universiteit Utrecht/CSDB/Major Internship/Modelling cell cycle/'

lambdaG1_DMSO_1st=0.74
kappaG1_DMSO_1st=10.5
lambdaS_DMSO_1st=18
kappaS_DMSO_1st=149
lambdaG2_DMSO_1st=13.2
kappaG2_DMSO_1st=41
lambdaM_DMSO_1st=35.5
kappaM_DMSO_1st=16
####
lambdaG1_V1_1st=0.74
kappaG1_V1_1st=10.5
lambdaS_V1_1st=15.35
kappaS_V1_1st=206.5
lambdaG2_V1_1st=13.2
kappaG2_V1_1st=41
lambdaM_V1_1st=35.5
kappaM_V1_1st=16

lambdaG1_DMSO_2nd=2.6
kappaG1_DMSO_2nd=12.5
lambdaS_DMSO_2nd=21.9
kappaS_DMSO_2nd=132.5
lambdaG2_DMSO_2nd=9.25
kappaG2_DMSO_2nd=29.5
lambdaM_DMSO_2nd=35.5
kappaM_DMSO_2nd=16
####
lambdaG1_V1_2nd=17.0
kappaG1_V1_2nd=120.5
lambdaS_V1_2nd=21.65
kappaS_V1_2nd=235.5
lambdaG2_V1_2nd=9.25
kappaG2_V1_2nd=29.5
lambdaM_V1_2nd=35.5
kappaM_V1_2nd=16

lambdaG1_DMSO_3rd=1.65
kappaG1_DMSO_3rd=13
lambdaS_DMSO_3rd=11.2
kappaS_DMSO_3rd=60.5
lambdaG2_DMSO_3rd=8.85
kappaG2_DMSO_3rd=17.5
lambdaM_DMSO_3rd=35.5
kappaM_DMSO_3rd=16
####
lambdaG1_V1_3rd=14.25
kappaG1_V1_3rd=88.0
lambdaS_V1_3rd=16.2
kappaS_V1_3rd=266.0
lambdaG2_V1_3rd=8.85
kappaG2_V1_3rd=17.5
lambdaM_V1_3rd=35.5
kappaM_V1_3rd=16

progression_1st=100
progression_2nd=100
stallingcells_DMSO_1st=20
stallingcells_DMSO_2nd=0
stallingcells_DMSO_3rd=0
stallingcells_V1_1st=20
stallingcells_V1_2nd=25
stallingcells_V1_3rd=25

errorcountDMSO=False
errorcountV1=False

dkappaandlambdaDMSO = {'G1 kappa':[kappaG1_DMSO_1st,kappaG1_DMSO_2nd,kappaG1_DMSO_3rd],
                        'G1 lambda':[lambdaG1_DMSO_1st,lambdaG1_DMSO_2nd,lambdaG1_DMSO_3rd],
                        'S kappa':[kappaS_DMSO_1st,kappaS_DMSO_2nd,kappaS_DMSO_3rd],
                        'S lambda':[lambdaS_DMSO_1st,lambdaS_DMSO_2nd,lambdaS_DMSO_3rd],
                        'G2 kappa':[kappaG2_DMSO_1st,kappaG2_DMSO_2nd,kappaG2_DMSO_3rd],
                        'G2 lambda':[lambdaG2_DMSO_1st,lambdaG2_DMSO_2nd,lambdaG2_DMSO_3rd],
                      'M kappa':[kappaM_DMSO_1st,kappaM_DMSO_2nd,kappaM_DMSO_3rd],
                      'M lambda':[lambdaM_DMSO_1st,lambdaM_DMSO_2nd,lambdaM_DMSO_3rd]}
dfkappaandlambdaDMSO=pd.DataFrame(data=dkappaandlambdaDMSO, index=['1st cell cycle', '2nd cell cycle','3rd cell cycle'])

dkappaandlambdaV1 = {'G1 kappa':[kappaG1_V1_1st,kappaG1_V1_2nd,kappaG1_V1_3rd],
                        'G1 lambda':[lambdaG1_V1_1st,lambdaG1_V1_2nd,lambdaG1_V1_3rd],
                        'S kappa':[kappaS_V1_1st,kappaS_V1_2nd,kappaS_V1_3rd],
                        'S lambda':[lambdaS_V1_1st,lambdaS_V1_2nd,lambdaS_V1_3rd],
                        'G2 kappa':[kappaG2_V1_1st,kappaG2_V1_2nd,kappaG2_V1_3rd],
                        'G2 lambda':[lambdaG2_V1_1st,lambdaG2_V1_2nd,lambdaG2_V1_3rd],
                    'M kappa':[kappaM_V1_1st,kappaM_V1_2nd,kappaM_V1_3rd],
                     'M lambda':[lambdaM_V1_1st,lambdaM_V1_2nd,lambdaM_V1_3rd]}
dfkappaandlambdaV1=pd.DataFrame(data=dkappaandlambdaV1, index=['1st cell cycle', '2nd cell cycle','3rd cell cycle'])

#Running model for DMSO treated cells

outputPalboDMSO=gettingstartpoints(numberofcells = numberofcells,
                                   lambdaG1 = lambdaG1_DMSO_1st,kappaG1 = kappaG1_DMSO_1st,lambdaS = lambdaS_DMSO_1st,kappaS = kappaS_DMSO_1st,lambdaG2 = lambdaG2_DMSO_1st,kappaG2 = kappaG2_DMSO_1st,lambdaM = lambdaM_DMSO_1st,kappaM = kappaM_DMSO_1st,
                                  outputpath=outputpath,treatment='DMSO')
dfPalboDMSO,errorcountDMSO=ReleaseIntoCellCycle4(outputPalboDMSO[0],outputPalboDMSO[1],outputPalboDMSO[2],outputPalboDMSO[3],outputPalboDMSO[4],outputPalboDMSO[5],outputPalboDMSO[6],
                                  lambdaG1_2nd = lambdaG1_DMSO_2nd,kappaG1_2nd = kappaG1_DMSO_2nd,lambdaS_2nd = lambdaS_DMSO_2nd,kappaS_2nd = kappaS_DMSO_2nd,lambdaG2_2nd = lambdaG2_DMSO_2nd,kappaG2_2nd = kappaG2_DMSO_2nd,lambdaM_2nd = lambdaM_DMSO_2nd,kappaM_2nd = kappaM_DMSO_2nd,
                                  lambdaG1_3rd = lambdaG1_DMSO_3rd,kappaG1_3rd = kappaG1_DMSO_3rd,lambdaS_3rd = lambdaS_DMSO_3rd,kappaS_3rd = kappaS_DMSO_3rd,lambdaG2_3rd = lambdaG2_DMSO_3rd,kappaG2_3rd = kappaG2_DMSO_3rd,lambdaM_3rd = lambdaM_DMSO_3rd,kappaM_3rd = kappaM_DMSO_3rd,
                                  progressionpercentage=progression_1st,progressionpercentage_2nd=progression_2nd,
                                  stallingcells_1st_G1=stallingcells_DMSO_1st,stallingcells_2nd_G1=stallingcells_DMSO_2nd,stallingcells_3rd_G1=stallingcells_DMSO_3rd,
                                  takingtimepoints=takingtimepoints,deltatime=deltatimeinterval,
                                  outputpath=outputpath,treatment='DMSO')

#Running model for dTAG-V1 treated cells
outputPalboV1=gettingstartpoints(numberofcells = numberofcells,
                                 lambdaG1 = lambdaG1_V1_1st,kappaG1 = kappaG1_V1_1st,lambdaS = lambdaS_V1_1st,kappaS = kappaS_V1_1st,lambdaG2 = lambdaG2_V1_1st,kappaG2 = kappaG2_V1_1st,lambdaM = lambdaM_V1_1st,kappaM = kappaM_V1_1st,
                                outputpath=outputpath,treatment='V1')
dfPalboV1,errorcountV1=ReleaseIntoCellCycle4(outputPalboV1[0],outputPalboV1[1],outputPalboV1[2],outputPalboV1[3],outputPalboV1[4],outputPalboV1[5],outputPalboV1[6],
                                lambdaG1_2nd = lambdaG1_V1_2nd,kappaG1_2nd = kappaG1_V1_2nd,lambdaS_2nd = lambdaS_V1_2nd,kappaS_2nd = kappaS_V1_2nd,lambdaG2_2nd = lambdaG2_V1_2nd,kappaG2_2nd = kappaG2_V1_2nd,lambdaM_2nd = lambdaM_V1_2nd,kappaM_2nd = kappaM_V1_2nd,
                                lambdaG1_3rd = lambdaG1_V1_3rd,kappaG1_3rd = kappaG1_V1_3rd,lambdaS_3rd = lambdaS_V1_3rd,kappaS_3rd = kappaS_V1_3rd,lambdaG2_3rd = lambdaG2_V1_3rd,kappaG2_3rd = kappaG2_V1_3rd,lambdaM_3rd = lambdaM_V1_3rd,kappaM_3rd = kappaM_V1_3rd,
                                progressionpercentage=progression_1st,progressionpercentage_2nd=progression_2nd,
                                stallingcells_1st_G1=stallingcells_V1_1st,stallingcells_2nd_G1=stallingcells_V1_2nd,stallingcells_3rd_G1=stallingcells_V1_3rd,
                                takingtimepoints=takingtimepoints,deltatime=deltatimeinterval,
                                outputpath=outputpath,treatment='V1')

realdata = pd.read_csv(str(inputpath+'ExperimentalData.txt'),sep='\t')
dfPalboV1temp=dfPalboV1.drop(columns=['time'])
concatdf = pd.concat([dfPalboDMSO, dfPalboV1temp], axis=1)
concatdf=concatdf.set_index('time')
concatdf=concatdf/numberofcells*100
concatdf.columns = ['G1 DMSO', 'S DMSO', 'G2M DMSO','G2 DMSO','M DMSO', 'G1 dTAG-V1','S dTAG-V1','G2M dTAG-V1','G2 dTAG-V1','M dTAG-V1']
concatdf['Time'] = concatdf.index

outputpathconcatdffile='C:/Users/Gebruiker/OneDrive - Universiteit Utrecht/CSDB/Major Internship/Modelling cell cycle/'
concatdf.to_csv(str(outputpathconcatdffile+'concatdf_model.txt'), header=True, index=True, sep='\t')

fig, axes = plt.subplots(1, 3)
fig.set_figwidth(18)
concatdf.iloc[:,[0,5]].plot(ax=axes[0],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),marker='o',
                           color={"G1 DMSO": "#5B9BD5", "G1 dTAG-V1": "#FF5050"})
concatdf.iloc[:,[1,6]].plot(ax=axes[1],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),marker='o',
                           color={"S DMSO": "#5B9BD5", "S dTAG-V1": "#FF5050"})
concatdf.iloc[:,[2,7]].plot(ax=axes[2],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),marker='o',
                           color={"G2M DMSO": "#5B9BD5", "G2M dTAG-V1": "#FF5050"})

fig.savefig(str(outputpath+'modellinegraph_lastmodel.png'))
plt.close()

fig, axes = plt.subplots(1, 3)
fig.set_figwidth(18)
concatdf.iloc[:,[0,5]].plot(ax=axes[0],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),
                           color={"G1 DMSO": "#5B9BD5", "G1 dTAG-V1": "#FF5050"})
realdata.plot.line(ax=axes[0],x='Time',y='G1 DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'],style=':')
realdata.plot.line(ax=axes[0],x='Time',y='G1 dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'],style=':')
realdata.plot.scatter(ax=axes[0],x='Time',y='G1 DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'])
realdata.plot.scatter(ax=axes[0],x='Time',y='G1 dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'])

concatdf.iloc[:,[1,6]].plot(ax=axes[1],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),
                           color={"S DMSO": "#5B9BD5", "S dTAG-V1": "#FF5050"})
realdata.plot.line(ax=axes[1],x='Time',y='S DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'],style=':')
realdata.plot.line(ax=axes[1],x='Time',y='S dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'],style=':')
realdata.plot.scatter(ax=axes[1],x='Time',y='S DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'])
realdata.plot.scatter(ax=axes[1],x='Time',y='S dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'])

concatdf.iloc[:,[2,7]].plot(ax=axes[2],ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),
                           color={"G2M DMSO": "#5B9BD5", "G2M dTAG-V1": "#FF5050"})
realdata.plot.line(ax=axes[2],x='Time',y='G2M DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'],style=':')
realdata.plot.line(ax=axes[2],x='Time',y='G2M dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'],style=':')
realdata.plot.scatter(ax=axes[2],x='Time',y='G2M DMSO exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#5B9BD5'])
realdata.plot.scatter(ax=axes[2],x='Time',y='G2M dTAG-V1 exp',ylim=[0,100],xlim=[0,48],xticks=list(range(0,49,3)),color=['#FF5050'])

fig.savefig(str(outputpath+'modellinegraph_withexperimentaldata_lastmodel.png'))
plt.close()

expectedgammadistributionsplotting(lambdaG1_DMSO_1st,kappaG1_DMSO_1st,lambdaS_DMSO_1st,kappaS_DMSO_1st,lambdaG2_DMSO_1st,kappaG2_DMSO_1st,lambdaM_DMSO_1st,kappaM_DMSO_1st,
                                       lambdaG1_V1_1st,kappaG1_V1_1st,lambdaS_V1_1st,kappaS_V1_1st,lambdaG2_V1_1st,kappaG2_V1_1st,lambdaM_V1_1st,kappaM_V1_1st,
                                       lambdaG1_DMSO_2nd,kappaG1_DMSO_2nd,lambdaS_DMSO_2nd,kappaS_DMSO_2nd,lambdaG2_DMSO_2nd,kappaG2_DMSO_2nd,lambdaM_DMSO_2nd,kappaM_DMSO_2nd,
                                       lambdaG1_V1_2nd,kappaG1_V1_2nd,lambdaS_V1_2nd,kappaS_V1_2nd,lambdaG2_V1_2nd,kappaG2_V1_2nd,lambdaM_V1_2nd,kappaM_V1_2nd,
                                       lambdaG1_DMSO_3rd,kappaG1_DMSO_3rd,lambdaS_DMSO_3rd,kappaS_DMSO_3rd,lambdaG2_DMSO_3rd,kappaG2_DMSO_3rd,lambdaM_DMSO_3rd,kappaM_DMSO_3rd,
                                       lambdaG1_V1_3rd,kappaG1_V1_3rd,lambdaS_V1_3rd,kappaS_V1_3rd,lambdaG2_V1_3rd,kappaG2_V1_3rd,lambdaM_V1_3rd,kappaM_V1_3rd,
                                  outputpath=outputpath)



### fitting Fourier serries for the slope.

### fitting Fourier serries for the slope.
def fourier_series(x, f, n=0):
    """
    Returns a symbolic fourier series of order `n`.

    :param n: Order of the fourier series.
    :param x: Independent variable
    :param f: Frequency of the fourier series
    """
    # Make the parameter objects for all the terms
    cos_a = parameters(','.join(['a{}'.format(i) for i in range(1, n + 1)]))
    sin_b = parameters(','.join(['b{}'.format(i) for i in range(1, n + 1)]))
    # Construct the series
    series = x*m+ c + sum(ai * cos(i * f * x) + bi * sin(i * f * x)
                             for i, (ai, bi) in enumerate(zip(cos_a, sin_b), start=1))
    return series

x, y = variables('x, y')
w,m,c, = parameters('w, m,c')
model_dictDMSO = {y: fourier_series(x, f=w, n=3)}

realdata=realdata.rename(columns={"G1 DMSO exp":"G1_DMSO_exp","G1 dTAG-V1 exp":"G1_dTAGV1_exp"})
realdataDMSO = realdata[realdata['Time'] >9]
realdataV1 = realdata[realdata['Time'] >12]

xdataDMSOexp=realdataDMSO['Time'].to_numpy()
xdataV1exp=realdataV1['Time'].to_numpy()
ydataDMSOexp=realdataDMSO['G1_DMSO_exp'].to_numpy()
ydataV1exp=realdataV1['G1_dTAGV1_exp'].to_numpy()

# Define a Fit object for this model and data
model_dictDMSO = {y: fourier_series(x, f=w, n=3)}
fitDMSOexp = Fit(model_dictDMSO,x=xdataDMSOexp, y=ydataDMSOexp)
fit_resultDMSOexp = fitDMSOexp.execute()
model_dictV1 = {y: fourier_series(x, f=w, n=3)}
fitV1exp = Fit(model_dictV1,x=xdataV1exp, y=ydataV1exp)
fit_resultV1exp = fitV1exp.execute()

fig, (ax1,ax2) = plt.subplots(1, 2)
fig.set_figwidth(15)
ax1.plot(xdataDMSOexp, ydataDMSOexp)
ax1.plot(xdataDMSOexp, fitDMSOexp.model(x=xdataDMSOexp, **fit_resultDMSOexp.params).y, ls=':')
ax1.set_ylim(0,100)
ax1.set_title('DMSO')
ax1.set_xlabel('Hours')
ax1.set_ylabel('% of cells')
ax2.plot(xdataV1exp, ydataV1exp)
ax2.plot(xdataV1exp, fitV1exp.model(x=xdataV1exp, **fit_resultV1exp.params).y, ls=':')
ax2.set_ylim(0,100)
ax2.set_title('dTAG-V1')
ax2.set_xlabel('Hours')
ax2.set_ylabel('% of cells')
fig.savefig(str(outputpath+'modelFourier_exp_lastmodel.png'))
plt.close()

fig, (ax1,ax2) = plt.subplots(1, 2)
fig.set_figwidth(15)
mvalueDMSO=fit_resultDMSOexp.value(m)
interceptDMSO=fit_resultDMSOexp.value(c)
ax1.plot(realdata['Time'], realdata["G1_DMSO_exp"])
ax1.plot([12,48],[12*mvalueDMSO+interceptDMSO,48*mvalueDMSO+interceptDMSO])
ax1.set_ylim(0,100)
ax1.set_xlim(0,48)
ax1.set_title('DMSO')
ax1.set_xlabel('Hours')
ax1.set_ylabel('% of cells')
mvalueV1=fit_resultV1exp.value(m)
interceptV1=fit_resultV1exp.value(c)
ax2.plot(realdata['Time'], realdata["G1_dTAGV1_exp"])
ax2.plot([15,48],[15*mvalueV1+interceptV1,48*mvalueV1+interceptV1])
ax2.plot()
ax2.set_ylim(0,100)
ax1.set_xlim(0,48)
ax2.set_title('dTAG-V1')
ax2.set_xlabel('Hours')
ax2.set_ylabel('% of cells')
fig.savefig(str(outputpath+'modelFourierwithslope_exp_lastmodel.png'))
plt.close()

##Doing the same thing for the model
modelDMSO=concatdf[concatdf['Time'] >11].iloc[::3,:]
modelV1=concatdf[concatdf['Time'] >14].iloc[::3,:]
xdataDMSOmodel=modelDMSO['Time'].to_numpy()
xdataV1model=modelV1['Time'].to_numpy()
ydataDMSOmodel=modelDMSO['G1 DMSO'].to_numpy()
ydataV1model=modelV1['G1 dTAG-V1'].to_numpy()
model_dictDMSO = {y: fourier_series(x, f=w, n=3)}
fitDMSOmodel = Fit(model_dictDMSO,x=xdataDMSOmodel, y=ydataDMSOmodel)
fit_resultDMSOmodel = fitDMSOmodel.execute()
model_dictV1 = {y: fourier_series(x, f=w, n=3)}
fitV1model = Fit(model_dictV1,x=xdataV1model, y=ydataV1model)
fit_resultV1model = fitV1model.execute()

fig, (ax1,ax2) = plt.subplots(1, 2)
fig.set_figwidth(15)
ax1.plot(xdataDMSOmodel, ydataDMSOmodel)
ax1.plot(xdataDMSOmodel, fitDMSOmodel.model(x=xdataDMSOmodel, **fit_resultDMSOmodel.params).y, ls=':')
ax1.set_ylim(0,100)
ax1.set_title('DMSO')
ax1.set_xlabel('Hours')
ax1.set_ylabel('% of cells')
ax2.plot(xdataV1model, ydataV1model)
ax2.plot(xdataV1model, fitV1model.model(x=xdataV1model, **fit_resultV1model.params).y, ls=':')
ax2.set_ylim(0,100)
ax2.set_title('dTAG-V1')
ax2.set_xlabel('Hours')
ax2.set_ylabel('% of cells')
fig.savefig(str(outputpath+'modelFourier_simulation_lastmodel.png'))
plt.close()

fig, (ax1,ax2) = plt.subplots(1, 2)
fig.set_figwidth(15)
mvalueDMSO=fit_resultDMSOmodel.value(m)
interceptDMSO=fit_resultDMSOmodel.value(c)
ax1.plot(concatdf['Time'], concatdf["G1 DMSO"])
ax1.plot([12,48],[12*mvalueDMSO+interceptDMSO,48*mvalueDMSO+interceptDMSO])
ax1.set_ylim(0,100)
ax1.set_xlim(0,48)
ax1.set_title('DMSO')
ax1.set_xlabel('Hours')
ax1.set_ylabel('% of cells')

mvalueV1=fit_resultV1model.value(m)
interceptV1=fit_resultV1model.value(c)
ax2.plot(concatdf['Time'], concatdf["G1 dTAG-V1"])
ax2.plot([15,48],[15*mvalueV1+interceptV1,48*mvalueV1+interceptV1])
ax2.plot()
ax2.set_ylim(0,100)
ax1.set_xlim(0,48)
ax2.set_title('dTAG-V1')
ax2.set_xlabel('Hours')
ax2.set_ylabel('% of cells')
fig.savefig(str(outputpath+'modelFourierwithslope_simulation_lastmodel.png'))
plt.close()





## Getting the R square as goodness of fit.
realdata = pd.read_csv(str(inputpath+'ExperimentalData.txt'),sep='\t')
SSresG1DMSO = 0
SStotG1DMSO = 0
SSresSDMSO = 0
SStotSDMSO = 0
SSresG2MDMSO = 0
SStotG2MDMSO = 0
SSresG1V1 = 0
SStotG1V1 = 0
SSresSV1 = 0
SStotSV1 = 0
SSresG2MV1 = 0
SStotG2MV1 = 0
SSresTotalDMSO=0
SSresTotalV1=0
SStotTotalDMSO=0
SStotTotalV1=0


modeldata=concatdf.iloc[::3,:]

G1modelDMSO=modeldata['G1 DMSO'].tolist()
G1expDMSO=realdata['G1 DMSO exp'].tolist()
G1modelV1=modeldata['G1 dTAG-V1'].tolist()
G1expV1=realdata['G1 dTAG-V1 exp'].tolist()
SmodelDMSO=modeldata['S DMSO'].tolist()
SexpDMSO=realdata['S DMSO exp'].tolist()
SmodelV1=modeldata['S dTAG-V1'].tolist()
SexpV1=realdata['S dTAG-V1 exp'].tolist()
G2MmodelDMSO=modeldata['G2M DMSO'].tolist()
G2MexpDMSO=realdata['G2M DMSO exp'].tolist()
G2MmodelV1=modeldata['G2M dTAG-V1'].tolist()
G2MexpV1=realdata['G2M dTAG-V1 exp'].tolist()


loop=0
for loop in range(0,len(G1expDMSO)):
    #print('i',loop)
    SSresG1DMSO+=(G1expDMSO[loop]-G1modelDMSO[loop])**2
    SStotG1DMSO+=(G1expDMSO[loop]-statistics.mean(G1expDMSO))**2
    SSresSDMSO+=(SexpDMSO[loop]-SmodelDMSO[loop])**2
    SStotSDMSO+=(SexpDMSO[loop]-statistics.mean(SexpDMSO))**2
    SSresG2MDMSO+=(G2MexpDMSO[loop]-G2MmodelDMSO[loop])**2
    SStotG2MDMSO+=(G2MexpDMSO[loop]-statistics.mean(G2MexpDMSO))**2
    SSresG1V1+=(G1expV1[loop]-G1modelV1[loop])**2
    SStotG1V1+=(G1expV1[loop]-statistics.mean(G1expV1))**2
    SSresSV1+=(SexpV1[loop]-SmodelV1[loop])**2
    SStotSV1+=(SexpV1[loop]-statistics.mean(SexpV1))**2
    SSresG2MV1+=(G2MexpV1[loop]-G2MmodelV1[loop])**2
    SStotG2MV1+=(G2MexpV1[loop]-statistics.mean(G2MexpV1))**2

    SSresTotalDMSO=SSresG1DMSO+SSresSDMSO+SSresG2MDMSO
    SStotTotalDMSO=SStotG1DMSO+SStotSDMSO+SStotG2MDMSO
    SSresTotalV1=SSresG1V1+SSresSV1+SSresG2MV1
    SStotTotalV1=SStotG1V1+SStotSV1+SStotG2MV1

print('R square G1 DMSO: ',1-(SSresG1DMSO/SStotG1DMSO))
print('R square S DMSO: ',1-(SSresSDMSO/SStotSDMSO))
print('R square G2 DMSO: ',1-(SSresG2MDMSO/SStotG2MDMSO))
print('R square total DMSO: ',1-(SSresTotalDMSO/SStotTotalDMSO))

print('R square G1 V1: ',1-(SSresG1V1/SStotG1V1))
print('R square S V1: ',1-(SSresSV1/SStotSV1))
print('R square G2M V1: ',1-(SSresG2MV1/SStotG2MV1))
print('R square total V1: ',1-(SSresTotalV1/SStotTotalV1))

from fpdf import FPDF
class PDF(FPDF):
     pass # nothing happens when it is executed.
outputpath='C:/Users/Gebruiker/OneDrive - Universiteit Utrecht/CSDB/Major Internship/Modelling cell cycle/TempFigures/'

###
pdf = PDF('P','mm','A4')#pdf object
pdf.set_margins(25, 25)
pdf.add_page()
pdf_w=210
pdf_h=297
pdf.set_font('helvetica','',24)
pdf.cell(40,10,str('Model and simulation '+str(modelnumber)),new_y='NEXT')
pdf.set_font('helvetica','',10)
pdf.cell(80,100,' ',new_y='NEXT',new_x='LMARGIN')
pdf.image(str(outputpath+'modellinegraph_lastmodel.png'),x=-5,y=40, w=220)
pdf.image(str(outputpath+'modellinegraph_withexperimentaldata_lastmodel.png'),x=-5,y=85, w=220)

dfDMSO = dfkappaandlambdaDMSO.applymap(str)
dfDMSO = dfDMSO.reset_index(level=0)
datacolumnsDMSO=dfDMSO.columns.values.tolist()
datavaluesDMSO=dfDMSO.values.tolist()
datavaluesDMSO.insert(0,datacolumnsDMSO)

pdf.cell(80,10,'\nKappa and lambda values DMSO samples',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width = pdf.epw / 9  # distribute content evenly
col_width=17
col_widthindex=25
for row in datavaluesDMSO:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)

dfV1 = dfkappaandlambdaV1.applymap(str)
dfV1 = dfV1.reset_index(level=0)
datacolumnsV1=dfV1.columns.values.tolist()
datavaluesV1=dfV1.values.tolist()
datavaluesV1.insert(0,datacolumnsV1)

pdf.set_font('helvetica','',10)
pdf.cell(80,5,' ',new_y='NEXT',new_x='LMARGIN')
pdf.cell(80,10,'Kappa and lambda values dTAG-V1 samples',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width=17
col_widthindex=25
for row in datavaluesV1:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)

data=[['','DMSO','dTAG-V1'],['1st cell cycle',str(stallingcells_DMSO_1st),str(stallingcells_V1_1st)],
 ['2nd cell cycle',str(stallingcells_DMSO_2nd),str(stallingcells_V1_2nd)],
 ['3rd cell cycle',str(stallingcells_DMSO_3rd),str(stallingcells_V1_3rd)]]
pdf.cell(80,5,' ',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=10)
pdf.cell(80,10,'Stalling in G1 after mitosis (%)',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width=20
col_widthindex=30
for row in data:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)

###
pdf.add_page()
slopedata=[[' ','DMSO slope G1','dTAG-V1 slope G1'],
           ['Fitted to experimental data',str(fit_resultDMSOexp.value(m)),str(fit_resultV1exp.value(m))],
           ['Fitted to modelled data', str(fit_resultDMSOmodel.value(m)), str(fit_resultV1model.value(m))]]
pdf.set_font("helvetica", size=10)
pdf.cell(80,10,'Slope G1 phase comparison model and experimental data',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width=40
col_widthindex=40
for row in slopedata:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)
pdf.cell(80,20,'',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=10)
pdf.cell(80,10,'R square values of the model',new_y='NEXT',new_x='LMARGIN')

Rsquaremodel=[[' ','R square DMSO','R square dTAG-V1'],
           ['G1 phase',str(1-(SSresG1DMSO/SStotG1DMSO)),str(1-(SSresG1V1/SStotG1V1))],
           ['S phase', str(1-(SSresSDMSO/SStotSDMSO)), str(1-(SSresSV1/SStotSV1))],
             ['G2M phase',str(1-(SSresG2MDMSO/SStotG2MDMSO)),str(1-(SSresG2MV1/SStotG2MV1))],
             ['Overall',str(1-(SSresTotalDMSO/SStotTotalDMSO)),str(1-(SSresTotalV1/SStotTotalV1))]]
pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width=40
col_widthindex=40
for row in Rsquaremodel:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)
pdf.cell(80,20,'',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=10)
pdf.cell(80,10,'R square values of the model',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
pdf.cell(80,10,'',new_y='NEXT',new_x='LMARGIN')

medianphaselength=[[' ','Median length DMSO','Median length dTAG-V1'],
           ['Release time',str((kappaG1_DMSO_1st/lambdaG1_DMSO_1st)*(1-(1/(3*kappaG1_DMSO_1st+0.2)))),str((kappaG1_V1_1st/lambdaG1_V1_1st)*(1-(1/(3*kappaG1_V1_1st+0.2))))],
           ['1st S phase', str((kappaS_DMSO_1st/lambdaS_DMSO_1st)*(1-(1/(3*kappaS_DMSO_1st+0.2)))), str((kappaS_V1_1st/lambdaS_V1_1st)*(1-(1/(3*kappaS_V1_1st+0.2))))],
             ['1st G2 phase',str((kappaG2_DMSO_1st/lambdaG2_DMSO_1st)*(1-(1/(3*kappaG2_DMSO_1st+0.2)))),str((kappaG2_V1_1st/lambdaG2_V1_1st)*(1-(1/(3*kappaG2_V1_1st+0.2))))],
             ['1st M phase',str((kappaM_DMSO_1st/lambdaM_DMSO_1st)*(1-(1/(3*kappaM_DMSO_1st+0.2)))),str((kappaM_V1_1st/lambdaM_V1_1st)*(1-(1/(3*kappaM_V1_1st+0.2))))],
                  ['2nd G1 phase',str((kappaG1_DMSO_2nd/lambdaG1_DMSO_2nd)*(1-(1/(3*kappaG1_DMSO_2nd+0.2)))),str((kappaG1_V1_2nd/lambdaG1_V1_2nd)*(1-(1/(3*kappaG1_V1_2nd+0.2))))],
                  ['2nd S phase',str((kappaS_DMSO_2nd/lambdaS_DMSO_2nd)*(1-(1/(3*kappaS_DMSO_2nd+0.2)))),str((kappaS_V1_2nd/lambdaS_V1_2nd)*(1-(1/(3*kappaS_V1_2nd+0.2))))],
                  ['2nd G2 phase',str((kappaG2_DMSO_2nd/lambdaG2_DMSO_2nd)*(1-(1/(3*kappaG2_DMSO_2nd+0.2)))),str((kappaG2_V1_2nd/lambdaG2_V1_2nd)*(1-(1/(3*kappaG2_V1_2nd+0.2))))],
                  ['2nd M phase',str((kappaM_DMSO_2nd/lambdaM_DMSO_2nd)*(1-(1/(3*kappaM_DMSO_2nd+0.2)))),str((kappaM_V1_2nd/lambdaM_V1_2nd)*(1-(1/(3*kappaM_V1_2nd+0.2))))],
                  ['3rd G1 phase',str((kappaG1_DMSO_3rd/lambdaG1_DMSO_3rd)*(1-(1/(3*kappaG1_DMSO_3rd+0.2)))),str((kappaG1_V1_3rd/lambdaG1_V1_3rd)*(1-(1/(3*kappaG1_V1_3rd+0.2))))],
                  ['3rd S phase',str((kappaS_DMSO_3rd/lambdaS_DMSO_3rd)*(1-(1/(3*kappaS_DMSO_3rd+0.2)))),str((kappaS_V1_3rd/lambdaS_V1_3rd)*(1-(1/(3*kappaS_V1_3rd+0.2))))],
                  ['3rd G2 phase',str((kappaG2_DMSO_3rd/lambdaG2_DMSO_3rd)*(1-(1/(3*kappaG2_DMSO_3rd+0.2)))),str((kappaG2_V1_3rd/lambdaG2_V1_3rd)*(1-(1/(3*kappaG2_V1_3rd+0.2))))],
                  ['3rd M phase',str((kappaM_DMSO_3rd/lambdaM_DMSO_3rd)*(1-(1/(3*kappaM_DMSO_3rd+0.2)))),str((kappaM_V1_3rd/lambdaM_V1_3rd)*(1-(1/(3*kappaM_V1_3rd+0.2))))]]

pdf.set_font("helvetica", size=8)
line_height = pdf.font_size * 2.5
col_width=40
col_widthindex=40
for row in medianphaselength:
    i=1
    for datum in row:
        if i ==1:
            pdf.multi_cell(col_widthindex, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        else:
            pdf.multi_cell(col_width, line_height, datum, border=1,
                    new_x="RIGHT", new_y="TOP", max_line_height=pdf.font_size)
        i+=1
    pdf.ln(line_height)
pdf.cell(80,20,'',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=10)
pdf.cell(80,10,'R square values of the model',new_y='NEXT',new_x='LMARGIN')
pdf.set_font("helvetica", size=8)
pdf.cell(80,10,'',new_y='NEXT',new_x='LMARGIN')





###
pdf.add_page()
pdf.set_font('helvetica','',16)
pdf.cell(40,10,'Gamma distributions: expected and simulated',new_y='NEXT',new_x='LMARGIN')
pdf.image(str(outputpath+'expectedgammadistributions_1DMSO_lastmodel.png'),x=10,y=40, h=40)
pdf.image(str(outputpath+'expectedgammadistributions_1V1_lastmodel.png'),x=10,y=80, h=40)
pdf.image(str(outputpath+'expectedgammadistributions_2DMSO_lastmodel.png'),x=10,y=120, h=40)
pdf.image(str(outputpath+'expectedgammadistributions_2V1_lastmodel.png'),x=10,y=160, h=40)
pdf.image(str(outputpath+'expectedgammadistributions_3DMSO_lastmodel.png'),x=10,y=200, h=40)
pdf.image(str(outputpath+'expectedgammadistributions_3V1_lastmodel.png'),x=10,y=240, h=40)

pdf.image(str(outputpath+'modelledgammadistributions_1DMSO_lastmodel.png'),x=100,y=40, h=40)
pdf.image(str(outputpath+'modelledgammadistributions_1V1_lastmodel.png'),x=100,y=80, h=40)
pdf.image(str(outputpath+'modelledgammadistributions_2DMSO_lastmodel.png'),x=100,y=120, h=40)
pdf.image(str(outputpath+'modelledgammadistributions_2V1_lastmodel.png'),x=100,y=160, h=40)
pdf.image(str(outputpath+'modelledgammadistributions_3DMSO_lastmodel.png'),x=100,y=200, h=40)
pdf.image(str(outputpath+'modelledgammadistributions_3V1_lastmodel.png'),x=100,y=240, h=40)

###
pdf.add_page()
pdf.set_font('helvetica','',16)
pdf.cell(40,10,'Number of S phases, mitosis and stalled cells',new_y='NEXT',new_x='LMARGIN')
pdf.set_font('helvetica','',12)
pdf.cell(80,10,'DMSO')
pdf.cell(40,10,'dTAG-V1',new_y='NEXT')
pdf.image(str(outputpath+'numberofmitosis_lastmodel_DMSO.png'),x=20,y=40, h=40)
pdf.image(str(outputpath+'numberofmitosis_lastmodel_V1.png'),x=100,y=40, h=40)
pdf.image(str(outputpath+'numberofSphases_lastmodel_DMSO.png'),x=20,y=80, h=40)
pdf.image(str(outputpath+'numberofSphases_lastmodel_V1.png'),x=100,y=80, h=40)
pdf.image(str(outputpath+'numberquiescentcellcycles_lastmodel_DMSO.png'),x=20,y=120, h=40)
pdf.image(str(outputpath+'numberquiescentcellcycles_lastmodel_V1.png'),x=100,y=120, h=40)

###
pdf.add_page()
pdf.set_font('helvetica','',16)
pdf.cell(40,10,'Modelling Fourier serries to get the slope of the curve',new_y='NEXT',new_x='LMARGIN')
pdf.set_font('helvetica','',12)
pdf.cell(80,10,str('Model formula:'),new_y='NEXT',new_x='LMARGIN')
pdf.set_font('helvetica','',8)
pdf.cell(20,5,str(str(model_dictDMSO[y])),new_y='NEXT',new_x='LMARGIN')
#pdf.cell(20,5,str(str(model_dictV1[y])),new_y='NEXT',new_x='LMARGIN')
pdf.set_font('helvetica','',10)
pdf.cell(40,7,' ',new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('Experimental data values:'),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('Goodness of fit by r quare of the DMSO: '+str(fit_resultDMSOexp.r_squared)),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('Goodness of fit by r quare of the dTAG-V1: '+str(fit_resultV1exp.r_squared)),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('m (slope) of the DMSO: '+str(fit_resultDMSOexp.value(m))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('m (slope) of the dTAG-V1: '+str(fit_resultV1exp.value(m))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('c (intercept) of the dTAG-V1: '+str(fit_resultDMSOexp.value(c))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('c (intercept) of the dTAG-V1: '+str(fit_resultV1exp.value(c))),new_y='NEXT',new_x='LMARGIN')
pdf.image(str(outputpath+'modelFourier_exp_lastmodel.png'),x=15,y=90, h=50)
pdf.image(str(outputpath+'modelFourierwithslope_exp_lastmodel.png'),x=15,y=140,h=50)

pdf.cell(40,80,' ',new_y='NEXT',new_x='LMARGIN')
pdf.add_page()
pdf.set_font('helvetica','',10)
pdf.cell(40,5,str('Simulated data values:'),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('Goodness of fit by r quare of the DMSO: '+str(fit_resultDMSOmodel.r_squared)),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('Goodness of fit by r quare of the dTAG-V1: '+str(fit_resultV1model.r_squared)),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('m (slope) of the DMSO: '+str(fit_resultDMSOmodel.value(m))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('m (slope) of the dTAG-V1: '+str(fit_resultV1model.value(m))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('c (intercept) of the dTAG-V1: '+str(fit_resultDMSOmodel.value(c))),new_y='NEXT',new_x='LMARGIN')
pdf.cell(40,5,str('c (intercept) of the dTAG-V1: '+str(fit_resultV1model.value(c))),new_y='NEXT',new_x='LMARGIN')
pdf.image(str(outputpath+'modelFourier_simulation_lastmodel.png'),x=15,y=60, h=50)
pdf.image(str(outputpath+'modelFourierwithslope_simulation_lastmodel.png'),x=15,y=110,h=50)



outputpathpdffile='C:/Users/Gebruiker/OneDrive - Universiteit Utrecht/CSDB/Major Internship/Modelling cell cycle/'
pdf.output(str(str(outputpathpdffile)+'Model_'+str(modelnumber)+'.pdf'))
print('made pdf')

print()
```
